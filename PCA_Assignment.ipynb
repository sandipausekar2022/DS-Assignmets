```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
wine=pd.read_csv('wine.csv')
```

```python
wine
```

```python
wine.info()
```

```python
wine.describe()
```

```python
wine[wine.duplicated()]
```

```python
wine2=wine.iloc[:,1:]
```

```python
wine2
```

```python
wine_ary=wine2.values
```

```python
wine_ary
```

```python
wine_norm=scale(wine_ary)
```

```python
wine_norm
```

```python
pca=PCA(n_components=13)
```

```python
pca_values=pca.fit_transform(wine_norm)
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var2=np.cumsum(np.round(var,decimals=4)*100)
var2
```

```python
pca.components_
```

```python
plt.plot(var2,color='red')
plt.xlabel('no.of PC')
plt.ylabel('cummilative variance')
```

```python
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
plt.figure(figsize=(15,13))
sns.scatterplot(data=final_df)
```

## Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,method='complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hclusters
```

```python
y_pred=hclusters.fit_predict(wine_norm)
y_pred
```

```python
wine['clusters']=y_pred
```

```python
wine
```

```python
wine['clusters'].value_counts()
```

## KMeans

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
    
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,11),wcss)
plt.xlabel('no. of clusters')
plt.ylabel('wcss')
plt.title('elbow graph')
```

```python
kclusters=KMeans(3,random_state=30).fit(wine_norm)
kclusters
```

```python
kclusters.labels_
```
