```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
sal=pd.read_csv('SalaryData_Train.csv')
sal
```

```python
sal.info()
```

```python
sal.describe()
```

```python
sal.isnull().sum()
```

```python
sal['workclass']=sal['workclass'].astype('category')
sal['education']=sal['education'].astype('category')
sal['maritalstatus']=sal['maritalstatus'].astype('category')
sal['occupation']=sal['occupation'].astype('category')
sal['relationship']=sal['relationship'].astype('category')
sal['race']=sal['race'].astype('category')
sal['native']=sal['native'].astype('category')
sal['sex']=sal['sex'].astype('category')
```

```python
sal
```

```python
sal.info()
```

```python
from sklearn import preprocessing 
label_encoder=preprocessing.LabelEncoder()
```

```python
sal['Salary']=label_encoder.fit_transform(sal['Salary'])
```

```python
sal
```

```python
sal['workclass'] = label_encoder.fit_transform(sal['workclass'])
sal['education'] = label_encoder.fit_transform(sal['education'])
sal['maritalstatus'] = label_encoder.fit_transform(sal['maritalstatus'])
sal['occupation'] = label_encoder.fit_transform(sal['occupation'])
sal['relationship'] = label_encoder.fit_transform(sal['relationship'])
sal['race'] = label_encoder.fit_transform(sal['race'])
sal['sex'] = label_encoder.fit_transform(sal['sex'])
sal['native'] = label_encoder.fit_transform(sal['native'])
```

```python
sal
```

```python
x=sal.iloc[:,0:13]
y=sal.iloc[:,13]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.3, random_state=0)
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
acc=accuracy_score(y_pred,y_test)*100
acc
```

```python
pd.crosstab(y_pred,y_test)
```

```python
y_pred
```
