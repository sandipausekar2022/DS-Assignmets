```python
 ## Test for Homoscedasticity or Heteroscedasticity 
```

```python
def get_standardized_value(vals):
    return (vals-vals.mean())/vals.std() 
```

```python
plt.scatter(get_standardized_value(model.fittedvalues),get_standardized_value(model.resid))
plt.title('Residual Plot')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
```

## improving the model

```python
startup_new=startup.drop(startup.index[[49]],axis=0).reset_index(drop=True)
```

```python
startup_new
```

# final model

```python
final_model=smf.ols('Profit~RDS+ADMS+MTKS',data=startup_new).fit()
```

```python
while final_model.rsquared<0.99:
    for c in [np.max(c)>1]:
        final_model=smf.ols('Profit~RDS+ADMS+MTKS',data=startup_new).fit()
        (c,_)=final_model.get_influence().cooks_distance
        c
        np.argmax(c),np.max(c)
        startup_new=startup_new.drop(startup_new.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        startup_new
    else:
        final_model=smf.ols('Profit~RDS+ADMS+MTKS',data=startup_new).fit()
        final_model.rsquared,final_model.aic
        print('Thus rsquared is improved to',final_model.rsquared)
    
```
