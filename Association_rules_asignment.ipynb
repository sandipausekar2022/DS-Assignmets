## Book Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
book=pd.read_csv('book.csv')
```

```python
book
```

```python
book.info()
```

```python
book.describe()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_items=apriori(book,min_support=0.1,use_colnames=True)
frequent_items
```

```python
rules=association_rules(frequent_items,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
rules[rules.lift>1]
```

```python
## scatter plot with 10% support and 70% confidence
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('scatter plot')
```

```python
frequent_items2=apriori(book,min_support=0.2,use_colnames=True)
frequent_items2
```

```python
rules2=association_rules(frequent_items2,metric='lift',min_threshold=0.5)
rules2
```

```python
## scatter plot with 20% support and 50% confidence
plt.scatter(rules2['support'],rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('scatter plot')
```

```python
frequent_items3=apriori(book,min_support=0.05,use_colnames=True)
frequent_items3
```

```python
rules3=association_rules(frequent_items3,metric='lift',min_threshold=0.8)
rules3
```

```python
# scatter plot with 5% support and 80% confidence
plt.scatter(rules3['support'],rules3['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('scatter plot')
```
