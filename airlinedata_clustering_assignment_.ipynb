## Hierarchical

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
airline=pd.read_csv('EastWestAirlines.csv')
```

```python
airline
```

```python
airline.info()
```

```python
airline.describe()
```

```python
airline[airline.duplicated()]
```

```python
airline=airline.drop(['ID#'],axis=1)
airline
```

```python
from sklearn.preprocessing import normalize
```

```python
airline_norm=pd.DataFrame(normalize(airline),columns=airline.columns)
```

```python
airline_norm
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(10,9))
dendrogram=sch.dendrogram(sch.linkage(airline_norm,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
```

```python
hc
```

```python
y_hc=hc.fit_predict(airline_norm)
```

```python
y_hc
```

```python
airline['clusters']=y_hc
```

```python
airline
```

```python
airline['clusters'].value_counts()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(airline['clusters'],airline['Balance'],c=y_hc)
```

## K-means

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
from sklearn.preprocessing import StandardScaler
```

```python
airline=pd.read_csv('EastWestAirlines.csv')
airline
```

```python
airline=airline.drop(['ID#'],axis=1)
airline
```

```python
airline_norm=StandardScaler().fit_transform(airline)
airline_norm
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(airline_norm)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('elbow curve')
plt.xlabel('No. of clusters')
plt.ylabel('wcss')
```

```python
hc=KMeans(4,random_state=40).fit(airline_norm)
```

```python
hc
```

```python
hc.labels_
```

```python
md=hc.labels_
```

```python
airline['clusters']=md
```

```python
airline
```

```python
airline.groupby('clusters').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(airline['clusters'],airline['Balance'],c=md)
```

## DBScan

```python
from sklearn.cluster import DBSCAN
```

```python
airline=pd.read_csv('EastWestAirlines.csv')
airline
```

```python
airline=airline.drop(['ID#'],axis=1)
airline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
airline_norm=StandardScaler().fit_transform(airline)
airline_norm
```

```python
dbscan=DBSCAN(eps=1,min_samples=11)
dbscan.fit(airline_norm)
```

```python
dbscan.labels_
```

```python
airline['clusterid']=dbscan.labels_
```

```python
airline
```

```python
airline.groupby('clusterid').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(airline['clusterid'],airline['Balance'],c=dbscan.labels_)
```

```python
airline['clusterid'].value_counts()
```
