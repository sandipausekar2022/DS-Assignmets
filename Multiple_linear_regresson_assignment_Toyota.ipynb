# model validation¶

# vif

```python
rsq_age=smf.ols('age~km+hp+cc+doors+gears+qt+wt',data=toyo).fit().rsquared
vif_age=1/(1-rsq_age)

rsq_km=smf.ols('km~age+hp+cc+doors+gears+qt+wt',data=toyo).fit().rsquared
vif_km=1/(1-rsq_km)

rsq_hp=smf.ols('hp~age+km+cc+doors+gears+qt+wt',data=toyo).fit().rsquared
vif_hp=1/(1-rsq_hp)

rsq_cc=smf.ols('cc~age+km+hp+doors+gears+qt+wt',data=toyo).fit().rsquared
vif_cc=1/(1-rsq_cc)

rsq_doors=smf.ols('doors~age+km+hp+cc+gears+qt+wt',data=toyo).fit().rsquared
vif_doors=1/(1-rsq_doors)

rsq_gears=smf.ols('gears~age+km+hp+cc+doors+qt+wt',data=toyo).fit().rsquared
vif_gears=1/(1-rsq_gears)

rsq_qt=smf.ols('qt~age+km+hp+cc+doors+gears+wt',data=toyo).fit().rsquared
vif_qt=1/(1-rsq_qt)

rsq_wt=smf.ols('wt~age+km+hp+cc+doors+gears+qt',data=toyo).fit().rsquared
vif_wt=1/(1-rsq_wt)

d={'colomns':['age','km','hp','cc','doors','gears','qt','wt'],'vif':[vif_age,vif_km,vif_hp,vif_cc,vif_doors,vif_gears,vif_qt,vif_wt]}
vif_df=pd.DataFrame(d)
vif_df
```

# improving the model

```python
toyo=toyo.drop(toyo.index[[80]],axis=0).reset_index(drop=True)
toyo
```

# final model

```python
while model.rsquared<0.90:
    for c in [np.max(c)>0.5]:
        model=smf.ols('Price~age+km+hp+cc+doors+gears+qt+wt',data=toyo).fit()
        (c,_)=model.get_influence().cooks_distance
        c
        np.argmax(c),np.max(c)
        toyo=toyo.drop(toyo.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        toyo
    else:
        model=smf.ols('Price~age+km+hp+cc+doors+gears+qt+wt',data=toyo).fit()
        model.rsquared , model.aic
        print('Thus rsquared is improved to',model.rsquared)
```
